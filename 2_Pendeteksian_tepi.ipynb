# Pendeteksian tepi menggunakan Gradien Prewitt

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Memuat citra dalam format grayscale
image = cv2.imread('/content/drive/MyDrive/PCD/luffy kasep.jpg', cv2.IMREAD_GRAYSCALE)

# 2. Definisikan kernel Prewitt
kernel_x = np.array([[-1, 0, 1],
                     [-1, 0, 1],
                     [-1, 0, 1]], dtype=np.float32)
kernel_y = np.array([[1, 1, 1],
                     [0, 0, 0],
                     [-1, -1, -1]], dtype=np.float32)

# 3. Aplikasikan konvolusi untuk gradien x dan y
grad_x = cv2.filter2D(image, -1, kernel_x)
grad_y = cv2.filter2D(image, -1, kernel_y)

# 4. Hitung magnitudo gradien
magnitude = cv2.magnitude(grad_x.astype(np.float32), grad_y.astype(np.float32))

# 5. Tampilkan hasil
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Gradient X (Horizontal)")
plt.imshow(grad_x, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Gradient Y (Vertical)")
plt.imshow(grad_y, cmap='gray')
plt.axis('off')

plt.figure()
plt.title("Edge Magnitude")
plt.imshow(magnitude, cmap='gray')
plt.axis('off')
plt.show()
```

  Operator Prewitt adalah salah satu teknik deteksi tepi yang sederhana namun efektif, digunakan untuk mendeteksi perubahan intensitas dalam citra digital. Dalam implementasinya, kernel Prewitt menghitung gradien intensitas dalam dua arah: horizontal (x) dan vertikal (y). Gradien horizontal digunakan untuk mendeteksi tepi yang sejajar dengan sumbu vertikal, sementara gradien vertikal mendeteksi tepi sejajar dengan sumbu horizontal. Setelah gradien dihitung, magnitudo gradien diperoleh untuk memadukan kedua informasi tersebut menjadi peta tepi yang komprehensif.

Langkah-langkah implementasi meliputi:

1. Memuat citra dalam format grayscale untuk fokus pada intensitas.
2. Mendefinisikan kernel Prewitt untuk arah horizontal dan vertikal.
3. Menerapkan konvolusi dengan kernel pada citra menggunakan fungsi cv2.filter2D.
4. Menggabungkan gradien horizontal dan vertikal dengan menghitung magnitudo menggunakan cv2.magnitude.
5. Menampilkan hasil berupa gradien horizontal, gradien vertikal, dan peta tepi keseluruhan.
